```python
from fuzzywuzzy import process

countries_merged_df = merged_df['Country'].unique()
countries_df_melted = df_melted['Country'].unique()

def map_country(country, country_list, threshold=90):
    match, score = process.extractOne(country, country_list)
    return match if score >= threshold else country

merged_df['Country'] = merged_df['Country'].apply(map_country, args=(countries_df_melted,))
print("Unique Country names after fuzzy matching in merged_df:")
print(merged_df['Country'].unique())
result_df = pd.merge(merged_df, df_melted, on=['Country', 'Year'], how='inner')
print("\nMerged DataFrame:")
result_df.columns
```

```python
columns_to_impute = result_df.columns[2:]

def impute_missing_values(df):
    for country in df['Country'].unique():
        country_df = df[df['Country'] == country]
        for col in columns_to_impute:
            for i in country_df.index:
                if pd.isna(df.at[i, col]):
                    current_year = df.at[i, 'Year']
                    if current_year == 2021:
                        prev_five_years = country_df.loc[country_df['Year'] < current_year, col].tail(5)
                        if prev_five_years.isna().all():
                            df.at[i, col] = 0
                        else:
                            df.at[i, col] = prev_five_years.mean()
                    else:
                        next_five_years = country_df.loc[country_df['Year'] > current_year, col].head(5)
                        if next_five_years.isna().all():
                            df.at[i, col] = 0
                        else:
                            df.at[i, col] = next_five_years.mean()

impute_missing_values(result_df)
```

```python
result_df['Surface Temperature'] = result_df['Temperature Index'].apply(
    lambda x: 'increase' if x > 0 else ('decrease' if x < 0 else 'unchange')
)
```

```python
# Select relevant columns for emissions of CO2, N2O, and CH4
emissions_columns = [col for col in result_df.columns if col.startswith('Emissions')]

# Create a new dataframe to store the total emissions of each gas
total_emissions_df = pd.DataFrame()

# Calculate the total emissions for each gas
total_emissions_df['Emission_CO2'] = result_df[[col for col in emissions_columns if 'CO2' in col]].sum(axis=1)
total_emissions_df['Emission_N2O'] = result_df[[col for col in emissions_columns if 'N2O' in col]].sum(axis=1)
total_emissions_df['Emission_CH4'] = result_df[[col for col in emissions_columns if 'CH4' in col]].sum(axis=1)

# Print the new dataframe
total_emissions_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Calculate the correlation matrix excluding 'Country', 'Year', and 'Surface Temperature'
correlation_matrix = result_df.drop(columns=['Country', 'Year', 'Surface Temperature']).corr()

# Get the correlation of 'Rice Production(Tonne)' with other features
rice_correlation = correlation_matrix['Rice Production(Tonne)'].drop(['Rice Production(Tonne)'])

# Select the top 10 features with the highest correlation
top_10_features = rice_correlation.abs().nlargest(10).index
top_10_correlation = rice_correlation[top_10_features]
print("Top 10 features correlated with Rice Production(Tonne):")
print(top_10_correlation)


# Plotting the correlations with vertical bars
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_correlation.index, y=top_10_correlation.values, palette='viridis')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Features')
plt.title('Top 10 Features Correlated with Rice Production(Tonne)')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability
plt.show()
```
